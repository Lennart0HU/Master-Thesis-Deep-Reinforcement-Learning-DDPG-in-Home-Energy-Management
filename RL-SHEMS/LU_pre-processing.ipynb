```python
# fill missing values using quadratic interpolation and plot filled values with surrounding

def plot_filled_values(df, i, k, id, col):
    timestamps1 = pd.to_datetime(df['timestamp'][df.loc[i-(k+10):i+10,col].index])
    timestamps2 = pd.to_datetime(df['timestamp'][range(i-k,i+1)])
    # Plotting
    plt.figure(figsize=(5, 3))
    plt.plot(timestamps1, df.loc[i-(k+10):i+10,col], color='blue', label='Interpolated values')                    
    plt.scatter(timestamps2, df.loc[i-k:i,col], color='red', label='Previously missing values')
    plt.legend()
    plt.title((id, col))
    plt.xticks(timestamps1, rotation=45)
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%y %H:%M'))
    plt.show()

for id in charging_dfs.keys():
    for col in ['e_production', 'e_consumption']:
        missing = charging_dfs[id][col].isnull()
        charging_dfs[id][col] = charging_dfs[id][col].interpolate(method='quadratic').clip(lower=0)

        if missing.sum() == 0:
            continue

        # Print indices of previously missing values and their new values
        print(f"charger_Id {id}, {col}: Missing values were found and filled at these indices:")

        k = 0
        indices = charging_dfs[id][missing].index

        for n, i in enumerate(indices):
            print(f"Index: {i}, Filled value: {charging_dfs[id].loc[i, col]}")

            if n < len(indices) - 1:
                if indices[n] + 1 == indices[n+1]:
                    k += 1
                else:
                    plot_filled_values(charging_dfs[id], i, k, id, col)
                    k = 0
            else:
                plot_filled_values(charging_dfs[id], i, k, id, col)


```

Calculation of EV SOC and time-connected:

For the lack of real ev soc data:
Assumption 1: Regarding the entire year, the largest amount charged in one single transaction, is likely charging the EV from near empty to full. This amount charged is assumed to be the usable ev battery capacity, ev_cap.
Assumption 2: At the end of each transaction the soc regarding the usable capacity is 100%. "Usable capacity" is not equal to the technical battery capacity. But rather the capacity that can be used. 

For each transaction: 1 - kwh of current transaction / ev_cap is the inferred soc at the start time of the transaction.

The inferred soc does not coincide with the technical soc of the ev battery. An example: a user might have unplugged the ev after it was charged from technical soc 60% up to 80%.  The inferred soc will begin at 80% and reach 100%. 
100% is not the fill of the battery, but the full amount of energy chargable in this transaction.

Limitation: In cases, where a smart mode was used in the data, and then the user unplugged the vehicle before reaching the full technical soc, the inferred soc will be too low. More energy could have been charged into the EV in the given transaction time.
It's impossible to know, how much additional energy could have been charged.



Alternative idea:
I could forget about socs and instead just define the "amount to be charged" e_T per transaction. This amount could be 0 for all e_charge == NaN.
For the first row of a transaction e_T == e_charger[transaction==T].sum(), so amount to be charged = total amount charged in the observed transaction.
This observed e_T would be reduced with every charging in every timestep to reach 0 by or before the last timestep of the transaction.



OR:
should I make even smarter assumptions about the soc:
whenever a transaction ends on 0 charging, while smartmode is off, or while there is surplus energy: I can safely assume that the ev battery is full.
    But what to do with all the transactions in which smart mode is on and there is not pv surplus? In those cases it'S possible that the battery is not full, but the charge simply stopped because the smart mode decided so.
        And what about all the transacitons not ending on 0? What soc would I assume here?

```python
for charger_id, charging_df in charging_dfs.items():
    print("charger_id: ", charger_id)

    charging_df['h_countdown'] = -1.0  # The countdown is the number of hours until the ev departs and ev_soc needs to be 100%. EV disconnected: -1

    charging_df['soc_ev'] = 1.0 # The soc_ev is the state of charge of the ev in % at the BEGINNING of this time step (so after the previous steps charging)

    e_transactions = []

    transaction_id_set = set(charging_df['transaction_id'].dropna())

    for transaction in transaction_id_set:
        e_transactions.append(charging_df[charging_df['transaction_id'] == transaction]['e_charger'].sum())
    
    ev_cap = max(e_transactions)
    print("ev capacity: ", ev_cap)

    for j, transaction in enumerate(transaction_id_set):
        
        transaction_length = len(charging_df[charging_df['transaction_id'] == transaction]['e_charger'])
        countdown = range(0, transaction_length)
        transaction_indeces = charging_df[charging_df['transaction_id'] == transaction].index

        for i, index in enumerate(transaction_indeces):
            if i == 0:
                charging_df.loc[index, "soc_ev"] = 1.0 - e_transactions[j] / ev_cap
            else:
                charging_df.loc[index, "soc_ev"] = charging_df.loc[transaction_indeces[i-1], "soc_ev"] + (charging_df.loc[transaction_indeces[i-1], "e_charger"] / ev_cap)
            
            charging_df.loc[index, "h_countdown"] = countdown[-(i+1)] / 4 # quarter hours left till ev departure
        
        
    
```
