```python
chargerIDs = ['Charger01', 'Charger02', 'Charger03', 'Charger04', 'Charger05', 'Charger06', 'Charger07', 'Charger08', 'Charger09', 'Charger98']
runs = ['train', 'eval', 'test']

for ID in chargerIDs:
    for run in runs:
        df = pd.read_csv(f"single_building/data/{ID}_all_{run}_fix.csv")
        print('NaNs:', df['chargekwh'].isnull().sum())
        for i in range(0, len(df) -1 ):
            c_ev = df.loc[i,'h_countdown']
            if c_ev > -1:
                if df.loc[i+1,'h_countdown'] != c_ev - 1:
                    print(f'ID: {ID}, run: {run}, index:', i)
                    print(c_ev)
                    print(df.loc[i+1,'h_countdown'])

```
